### 4.1 Variables: Storing Information

Variables are like labeled boxes that store values:

```python
# Storing numbers
mars_distance_km = 225_000_000  # Distance from Earth to Mars (avg)
speed_of_light = 299_792  # km/s

# Storing text (called "strings")
mission_name = "Perseverance"
planet = "Mars"

# You can display values with print()
print(f"The {mission_name} rover is exploring {planet}")
print(f"Average distance to {planet}: {mars_distance_km:,} km")
```

### 4.2 Doing Math

Python can be used as a powerful calculator:

```python
# Basic operations
total_mass = 1000 + 500  # Addition
fuel_remaining = 1000 - 300  # Subtraction
orbital_period = 90 * 60  # Multiplication (90 minutes in seconds)
average_speed = 1000 / 10  # Division

# Powers (like for rocket equation!)
escape_velocity_squared = 11.2 ** 2

print(f"Spacecraft mass: {total_mass} kg")
print(f"Fuel remaining: {fuel_remaining} kg")
print(f"Orbital period: {orbital_period} seconds")
print(f"Escape velocity squared: {escape_velocity_squared} km¬≤/s¬≤")
```

### 4.3 Lists: Storing Multiple Values

Lists let you store many related values together:

```python
# A list of planets
inner_planets = ["Mercury", "Venus", "Earth", "Mars"]

# A list of distances from the Sun (in AU)
distances_from_sun = [0.39, 0.72, 1.0, 1.52]

# Accessing items (counting starts at 0!)
print(f"First planet: {inner_planets[0]}")
print(f"Fourth planet: {inner_planets[3]}")
print(f"Mars is {distances_from_sun[3]} AU from the Sun")

# How many items?
print(f"\nNumber of inner planets: {len(inner_planets)}")
```

### 4.4 Loops: Doing Things Repeatedly

Loops let you repeat actions without writing the same code over and over:

```python
# Loop through all planets
print("The inner planets:")
for planet in inner_planets:
    print(f"  - {planet}")

print("\nPlanets with their distances:")
for i in range(len(inner_planets)):
    print(f"  - {inner_planets[i]}: {distances_from_sun[i]} AU from Sun")
```

### 4.5 Libraries: Using Pre-written Code

Python has thousands of libraries with useful functions. We'll use several throughout the course:

```python
# Import the math library
import math

# Use functions from the library
radius = 6371  # Earth's radius in km
circumference = 2 * math.pi * radius

print(f"Earth's circumference: {circumference:.0f} km")
print(f"Square root of 16: {math.sqrt(16)}")
```

### 4.6 Making Graphs

## Part 6: Practice Exercise - Exploring Space Mission Distances

Let's put everything together with a space-themed exercise!

### Scenario

You're planning communication with several spacecraft exploring the solar system. Radio signals travel at the speed of light, but even at that incredible speed, there's a noticeable delay when communicating with distant spacecraft.

### Your Task

Calculate the communication delay (round-trip) for several spacecraft at their current distances from Earth.

**Run the cell below to see the spacecraft data:**

```python
# Spacecraft and their approximate distances from Earth (as of 2026)
spacecraft_data = {
    "ISS": 408,                    # km
    "Moon (Apollo site)": 384_400,  # km
    "Mars Perseverance": 225_000_000,  # km (varies with orbital position)
    "Voyager 1": 24_000_000_000,   # km (in interstellar space!)
}

# Speed of light
speed_of_light_km_per_sec = 299_792  # km/s

print("Spacecraft distances from Earth:")
for spacecraft, distance in spacecraft_data.items():
    print(f"  {spacecraft}: {distance:,} km")
```

### Exercise 6.1: Calculate One-Way Signal Time

**Calculate how long it takes for a signal to reach the ISS from Earth:**

Formula: `time = distance / speed`

Fill in the code below:

```python
# Get the distance to ISS
iss_distance = spacecraft_data["ISS"]

# Calculate the time (in seconds)
# YOUR CODE HERE - calculate: iss_distance / speed_of_light_km_per_sec
signal_time_to_iss = iss_distance / speed_of_light_km_per_sec

# Display the result
print(f"Signal takes {signal_time_to_iss:.4f} seconds to reach the ISS")
print(f"That's {signal_time_to_iss * 1000:.2f} milliseconds!")
```

### Exercise 6.1b: ISS at Maximum Distance

**The ISS orbits at approximately 408 km altitude, but that's the closest it gets!**

When the ISS is on the opposite side of Earth from you, the signal has to travel:
- Up to the ISS (408 km)
- PLUS across Earth's diameter

Earth's radius is approximately 6,371 km.

**Calculate the maximum distance to the ISS and the signal time at that distance:**

```python
# Earth's radius
earth_radius_km = 6371

# ISS orbital altitude
iss_altitude = 408

# YOUR CODE HERE:
# Calculate the maximum distance when ISS is on the opposite side of Earth
# Hint: You need to go up to orbit, across Earth's diameter, then up to orbit again
# Maximum distance = ?
max_iss_distance = earth_radius_km*2 + iss_altitude*2

# Calculate signal time at maximum distance
# YOUR CODE HERE
max_signal_time = max_iss_distance / speed_of_light_km_per_sec

# Compare with minimum distance
min_iss_distance = iss_altitude
min_signal_time = min_iss_distance / speed_of_light_km_per_sec

print(f"ISS closest distance: {min_iss_distance:,} km")
print(f"ISS furthest distance: {max_iss_distance:,} km")
print(f"\nSignal time (closest): {min_signal_time*1000:.2f} milliseconds")
print(f"Signal time (furthest): {max_signal_time*1000:.2f} milliseconds")
print(f"\nDifference: {(max_signal_time - min_signal_time)*1000:.2f} milliseconds")
```

### Exercise 6.2: Calculate Round-Trip Times for All Spacecraft

**Now calculate the round-trip delay for all spacecraft**

(Signal goes to spacecraft, then the reply comes back = double the one-way time)

Complete the code below:

```python
print("Round-trip communication delays:\n")

# Loop through all spacecraft
for spacecraft, distance in spacecraft_data.items():
    # Calculate one-way time
    # YOUR CODE HERE
    one_way_time = (distance) / speed_of_light_km_per_sec

    # Calculate round-trip time (multiply by 2)
    # YOUR CODE HERE
    round_trip_time = one_way_time*2

    # Display in appropriate units
    if round_trip_time < 1:
        print(f"{spacecraft:20s}: {round_trip_time*1000:.2f} milliseconds")
    elif round_trip_time < 60:
        print(f"{spacecraft:20s}: {round_trip_time:.2f} seconds")
    elif round_trip_time < 3600:
        print(f"{spacecraft:20s}: {round_trip_time/60:.2f} minutes")
    else:
        print(f"{spacecraft:20s}: {round_trip_time/3600:.2f} hours")
```

### Exercise 6.3: Visualize the Data

**Create a bar chart showing the round-trip times**

We'll help you get started:

```python
round_trip_time
```

```python
import matplotlib.pyplot as plt

# Calculate round-trip times and store them
spacecraft_names = []
round_trip_times_hours = []

for spacecraft, distance in spacecraft_data.items():
    one_way_time = distance / speed_of_light_km_per_sec
    round_trip_time = one_way_time*2

    spacecraft_names.append(spacecraft)
    round_trip_times_hours.append(round_trip_time / 3600)  # Convert to hours

# Create the plot
plt.figure(figsize=(12, 6))

# YOUR CODE HERE: Create a bar chart
plt.bar(spacecraft_names, round_trip_times_hours, color=['green', 'yellow', 'blue', 'red'])

# Add labels and title
plt.ylabel('Round-trip signal time (hours)', fontsize=12)
plt.xlabel('Spacecraft', fontsize=12)
plt.title('Communication Delays with Spacecraft in the Solar System', fontsize=14, fontweight='bold')
# YOUR CODE HERE: The values vary enormously - some are milliseconds, others are hours!
# How can you display them all clearly on one chart?
plt.grid(axis='y', alpha=0.3)
plt.yscale('log')

plt.tight_layout()
plt.show()
```
